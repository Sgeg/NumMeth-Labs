```python
import numpy as np
import sympy as sp
from sympy.plotting import plot
from sympy.parsing.sympy_parser import parse_expr
import matplotlib.pyplot as plt
```

```python
sp.init_printing(use_latex = 'mathjax')
from IPython.display import display
```

```python
def getPoints(f, l, r, n):
    delta = (r - l) / (n - 1)
    points = []
    for i in range(n):
        points.append((l + i * delta, f.subs(list(f.free_symbols)[0], l + i * delta)))
    return points
```

```python
def parsePoints(s):
    points = []
    pairs = s.split(';')
    for i in range(len(pairs) - 1):
        fun = pairs[i].split()
        points.append((float(fun[0]), float(fun[1])))
    return (points, int(pairs[-1]))
```

```python
def gauss(B):
    A = B.copy()
    if not A.shape[0] + 1 == A.shape[1]:
        return 'Wrong matrix shape'
    for i in range(0, A.shape[0]):
        maxEl = np.absolute(A)[i:,i].argmax() + i
        A[[i, maxEl]] = A[[maxEl, i]]
        if abs(A[i, i]) < 1e-6:
            return 'Matrix rank less than n'
        for j in range(i + 1, A.shape[0]):
            coef = -A[j, i] / A[i, i]
            for k in range(i, A.shape[1]):
                A[j, k] = A[j, k] + coef*A[i, k]
    for i in range(A.shape[0] - 1, -1, -1):
        for j in range(i - 1, -1, -1):
            coef = -A[j, i] / A[i, i]
            A[j, i] = 0
            A[j, -1] = A[j, -1] + coef * A[i, -1]
        A[i, -1] = A[i, -1] / A[i, i]
        A[i, i] = 1
    return A[:, -1]
```

```python
def lagrange(f):
    pol = np.zeros(len(f), dtype = 'float')
    allMult = np.zeros(len(f) + 1)
    allMult[0] = 1
    for i in range(len(f)):
        allMultPr = allMult.copy()
        for j in range(len(allMult)):
            allMult[j] = -allMult[j] * f[i][0]
        for j in range(len(allMult) - 1):
            allMult[j + 1] = allMult[j + 1] + allMultPr[j]
    for i in range(len(f)):
        res = np.zeros(len(f))
        curr = allMult.copy()
        for j in range(len(curr) - 1, 0, -1):
            curr[j - 1] = curr[j - 1] + curr[j] * f[i][0]
            res[j - 1] = curr[j]
        coef = f[i][1]
        for j in range(len(f)):
            if(i != j):
                coef = coef / (f[i][0] - f[j][0])
        for j in range(len(res)):
            pol[j] = pol[j] + res[j] * coef
    ind = len(pol)
    for i in range(len(pol) - 1, 0):
        if(pol[i] != 0):
            break
        else:
            ind = i
    return pol[:ind]
```

# Method of Least Squares

```python
def leastSquares(points, n):
    x, y = map(list, zip(*points))
    
    xPow = [1] * len(x)
    xSum = [float] * (2*n + 1)
    for i in range(len(xSum)):
        xSum[i] = sum(xPow)
        xPow = [xPow[i] * x[i] for i in range(len(x))]
    
    yPow = y.copy()
    ySum = [float] * (n + 1)
    for i in range(len(ySum)):
        ySum[i] = sum(yPow)
        yPow = [yPow[i] * x[i] for i in range(len(x))]
    
    M = np.empty([n + 1, n + 2])
    M[:, -1] = ySum
    for i in range(M.shape[0]):
        for j in range(M.shape[1] - 1):
            M[i, j] = xSum[i + j]
    res = gauss(M)
    if type(res) is str:
        return(lagrange(points))
    return(res)
```

```python
def toPolynom(coef):
    x = sp.symbols('x')
    expr = sp.Float(coef[0])
    for i in range(1, len(coef)):
        expr = expr + coef[i] * x**i
    return expr
```

```python
with open('input.txt') as f:
    lines = f.readlines()
for i in range(len(lines)):
    points, n = parsePoints(lines[i])
    coef = leastSquares(points, n)
    pol = toPolynom(coef)
    
    points = sorted(points)
    x, y = map(list, zip(*points))
    l, r = points[0][0], points[-1][0]
    delta = 0.2
    val, image = map(list, zip(*getPoints(pol, l - delta, r + delta, int(100*(r - l + 2*delta)))))
    plt.plot(val, image, 'r', x, y, 'bo')
    display(pol)
    plt.show()
```

```python
with open('inputFunc.txt') as f:
    fileData = f.read()
lines = fileData.split(';')
for l in lines:
    line = l.split('\n')
    line = [x for x in line if x]
    expr = parse_expr(line[0])
    nums = line[1].split()
    l = float(nums[0])
    r = float(nums[1])
    num = int(nums[2])
    n = int(line[2])

    points = getPoints(expr, l, r, num)
    px, py = map(list, zip(*points))
    pol = toPolynom(leastSquares(points, n))
    delta = 0.2
    x, y = map(list, zip(*getPoints(expr, l - delta, r + delta, int(100*(r - l)))))
    polx, poly = map(list, zip(*getPoints(pol, l - delta, r + delta, int(100*(r - l)))))
    display(expr)
    display(pol)
    plt.plot(x, y, polx, poly, 'r--', px, py, 'ro')
    plt.show()
    print()
```
